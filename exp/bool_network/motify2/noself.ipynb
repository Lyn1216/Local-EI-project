```python
order = ['A', 'AB', 'AC', 'ABC']
```

```python
directory = Path('FFL/POS')
candi_sys = [0, 1, 2]

for file_path in directory.rglob('*'):
    if file_path.is_file():
        print(file_path)
        _,un_pos_result[str(file_path)[8:-8]],syn_pos_result[str(file_path)[8:-8]],ei_pos_result[str(file_path)[8:-8]],_,_,_,_ =\
        text_bn_graph(textfile=str(file_path), candidate_sys=candi_sys, fill_onenode=True, noise=0.1, save_onenote=False)
```

```python
un_pos_result
```

```python
average_un_pos_result = {}
for num in range(1,9):
    all_un_pos = 0
    for suffix in order:
        all_un_pos += un_pos_result[f'{num}_{suffix}']
        average_un_pos_result[f'type_{num}'] = all_un_pos / 4
average_un_pos_result
```

```python
keys = list(average_un_pos_result.keys())
values = list(average_un_pos_result.values())
plt.figure(figsize=(12, 6))
plt.plot(keys, values, label='un_pos', marker='o', markersize=3) 
plt.title('average un pos Result Line Chart')
plt.xlabel('Keys')
plt.ylabel('Values')
plt.subplots_adjust(bottom=0.2) 
plt.xticks(keys, rotation=45) 
plt.legend()
#plt.savefig('average_un_pos_result_line_chart.png', format='png', dpi=300)
plt.show()
```
