```python
import sys
sys.path.append("..")
import load_database13 as db
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from grn.grn_tpm import text_bn_graph
from pathlib import Path
```

```python
un_result = {}
syn_result = {}
ei_result = {}
```

```python
directory = Path('FFL')
candi_sys = [0, 1, 2]

for file_path in directory.rglob('*'):
    if file_path.is_file():
        print(file_path)
        _,un_result[str(file_path)[4:-4]],syn_result[str(file_path)[4:-4]],ei_result[str(file_path)[4:-4]],_,_,_,_=\
        text_bn_graph(textfile=str(file_path), candidate_sys=candi_sys, fill_onenode=True, noise=0, save_onenote=True)
```

```python
un_result
```

```python
un = {}
order = ['A', 'B', 'C', 'ABC']
for num in range(1, 9): 
    for suffix in order:
        un[f'{num}_{suffix}'] = un_result[f'{num}_{suffix}']
un
```

```python
syn_result
```

```python
syn = {}
order = ['A', 'B', 'C', 'ABC']
for num in range(1, 9): 
    for suffix in order:
        syn[f'{num}_{suffix}'] = syn_result[f'{num}_{suffix}']
syn
```

```python
ei_result
```

```python
ei = {}
order = ['A', 'B', 'C', 'ABC']
for num in range(1, 9): 
    for suffix in order:
        ei[f'{num}_{suffix}'] = ei_result[f'{num}_{suffix}']
ei
```

```python
average_syn_result = {}
order = ['A', 'B', 'C', 'ABC']
for num in range(1,9):
    all_syn = 0
    for suffix in order:
        all_syn += syn[f'{num}_{suffix}']
        average_syn_result[f'type_{num}'] = all_syn / 4
average_syn_result
```

```python
keys = list(average_syn_result.keys())
values = list(average_syn_result.values())
plt.figure(figsize=(12, 6))
plt.plot(keys, values, label='syn', marker='o', markersize=3) 
plt.title('average syn Result Line Chart')
plt.xlabel('Keys')
plt.ylabel('Values')
plt.subplots_adjust(bottom=0.2) 
plt.xticks(keys, rotation=45) 
plt.legend()
plt.savefig('average_syn_result_line_chart.png', format='png', dpi=300)
plt.show()
```

```python
keys = list(un.keys())
un_values = list(un.values())
syn_values = list(syn.values())
ei_values = list(ei.values())
plt.figure(figsize=(12, 6))
plt.plot(keys, un_values, label='un', marker='o', markersize=3) 
plt.plot(keys, syn_values, label='syn', marker='o', markersize=3) 
plt.plot(keys, ei_values, label='ei', marker='o', markersize=3) 
plt.title('Result Line Chart')
plt.xlabel('Keys')
plt.ylabel('Values')
plt.subplots_adjust(bottom=0.2) 
plt.xticks(keys, rotation=45) 
plt.legend()
plt.savefig('result_line_chart.png', format='png', dpi=300)
plt.show()
```
